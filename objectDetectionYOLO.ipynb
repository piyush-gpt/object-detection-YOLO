```python
from ultralytics import YOLO
```

```python
import cv2
import numpy as np
```

```python
my_file=open("C:\\Users\\lenovo\\Downloads\\coco.txt","r")
data=my_file.read()
class_list=data.split("\n")
my_file.close()
print(class_list)
```

```python
model=YOLO("weights/yolov8n.pt","v8")
```

```python
cap=cv2.VideoCapture(0)
frame_wid=640
frame_hyt=480
if not cap.isOpened():
    print("cannot open camera")
    exit()
```

```python
cv2.namedWindow("Resized_Window", cv2.WINDOW_NORMAL) 
cv2.resizeWindow("Resized_Window", 1000, 1000) 
cap=cv2.VideoCapture(0)
while True:
    ret, frame=cap.read()
    frame=cv2.flip(frame,1)
    if not ret:
        print("cannot retrive frame")
        break
    detect_params=model.predict(frame,conf=0.45,save=False)
    detect_params=detect_params[0].numpy()
    if len(detect_params) != 0:
        for j in range(len(detect_params.boxes)):
            box = detect_params.boxes[j]
            clsID = box.cls[0]
            conf = box.conf[0]
            bb = box.xyxy[0]
            cv2.rectangle(
                    frame,
                    (int(bb[0]), int(bb[1])),
                    (int(bb[2]), int(bb[3])),
                    (0,255,0),
                    3,
            )
    
                # Display class name and confidence
            font = cv2.FONT_HERSHEY_COMPLEX
            cv2.putText(
                    frame,
                    class_list[int(clsID)]
                    + " "
                    + str(round(conf, 3))
                    + "%",
                    (int(bb[0]), int(bb[1]) - 10),
                    font,
                    1,
                    (255, 255, 255),
                    2,
           )                

    # Display the resulting frame
    cv2.imshow('Resized_Window', frame)

    # Terminate run when "Q" pressed
    if cv2.waitKey(1) == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
